```python
import json
import os
import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

from Processing import Processor
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
import subprocess
subprocess.call(['./pdf_converter'])
```

```python
path = 'txt_files/'

files = []
# r=root, d=directories, f = files
for r, d, f in os.wal-k(path):
    for file in f:
        if '.txt' in file:
            files.append(os.path.join(r, file))
            
files[:10]
```

```python
processor = Processor(nlp)

storage = {}

for f in files:
    file = f[len(path):-4]
    storage[file] = processor.process(f, export=False)
```

```python
for file in storage.keys():
    print(set(storage[file]['institutions']))
```

```python
with open('papers' + '.json', 'w') as outfile:
    json.dump(storage, outfile)
```
